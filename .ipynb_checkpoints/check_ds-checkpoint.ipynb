```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('survey_results_public_mega_inc.csv')
df.drop(df.loc[ df['ConvertedSalary'] == 0.0 ].index , inplace=True)
df['LnConvertedSalary'] = np.log(df['ConvertedSalary'])
df
```

```python
df.columns.values.tolist()
```

```python
male = df.loc[df.Gender == 'Male']
female = df.loc[df.Gender == 'Female']
print("Male:\t ", len(male))
print("Female:\t ", len(female))
print("Other:\t ", len(df) - len(male) - len(female))
```

```python
white = df.loc[df.RaceEthnicity == 'White or of European descent']
black = df.loc[df.RaceEthnicity == 'Black or of African descent']
print("White:\t ", len(white))
print("Black:\t ", len(black))
print("Other:\t ", len(df) - len(white) - len(black))
```

```python
a = df.loc[df.FormalEducation == 'Bachelor’s degree (BA, BS, B.Eng., etc.)']
b = df.loc[df.FormalEducation == 'Some college/university study without earning a degree']
c = df.loc[df.FormalEducation == 'Master’s degree (MA, MS, M.Eng., MBA, etc.)']
d = df.loc[df.FormalEducation == 'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)']
e = df.loc[df.FormalEducation == 'I never completed any formal education']
f = df.loc[df.FormalEducation == 'Associate degree']
g = df.loc[df.FormalEducation == 'Other doctoral degree (Ph.D, Ed.D., etc.)']


print("BA:\t\t ", len(a))
print("MA:\t\t ", len(c))
print("Some College:\t ", len(b))
print("Secondary:\t ", len(d))
print("Associate:\t ", len(f))
print("Not complete:\t ", len(e))
print("Other:\t\t ", len(g))
```

```python
print(df.JobSatisfaction.unique())
df['JobSatisfaction'].replace({'Extremely satisfied': 7,
                                                    'Moderately satisfied': 6,
                                                    'Slightly satisfied': 5,
                                                    'Neither satisfied nor dissatisfied': 4,
                                                    'Slightly dissatisfied': 3,
                                                    'Moderately dissatisfied': 2,
                                                    'Extremely dissatisfied': 1}, inplace=True)
print(df.JobSatisfaction.unique())
```

```python
df.CareerSatisfaction.unique()
df['CareerSatisfaction'].replace({'Extremely satisfied': 7,
                                                    'Moderately satisfied': 6,
                                                    'Slightly satisfied': 5,
                                                    'Neither satisfied nor dissatisfied': 4,
                                                    'Slightly dissatisfied': 3,
                                                    'Moderately dissatisfied': 2,
                                                    'Extremely dissatisfied': 1}, inplace=True)
```

```python
#print(df.Dependents.unique())
df["Dependents"].replace({'Yes': 1,
                          'No': 0}, inplace=True)
print(df.Dependents.unique())
```

```python
print(df.YearsCoding.unique())
df["YearsCoding"].replace({'0-2 years': 2,
                           '3-5 years': 5,
                           '6-8 years': 8,
                           '9-11 years': 11,
                           '12-14 years': 14,
                           '15-17 years': 17,
                           '18-20 years': 20,
                           '21-23 years': 23,
                           '24-26 years': 26,
                           '27-29 years': 29,
                           '30 or more years': 30}, inplace=True)
df
```

```python
df.drop(df.columns.difference(['RaceEthnicity', 'JobSatisfaction', 'CareerSatisfaction', 'SexualOrientation','LnConvertedSalary','ConvertedSalary','YearsCoding','Gender', 'Dependents']), 1, inplace=True)
df = df.dropna()
df
```
